```python
from collections import defaultdict
from typing import List

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # Step 1: Build the graph
        graph = defaultdict(list)
        for course, prereq in prerequisites:
            graph[course].append(prereq)
 
        # Step 2: Define state constants
        UNVISITED = 0
        VISITING = 1
        VISITED = 2

        # State of each course
        state = [UNVISITED] * numCourses

        # Step 3: DFS to detect cycles
        def dfs(node):
            if state[node] == VISITING:
                return False  # Found a cycle
            if state[node] == VISITED:
                return True   # Already checked

            state[node] = VISITING  # Mark as visiting

            # Recur for all prerequisites
            for neighbor in graph[node]:
                if not dfs(neighbor):
                    return False

            state[node] = VISITED  # Mark as visited
            return True

        # Step 4: Check all courses
        for i in range(numCourses):
            if not dfs(i):
                return False

        return True
```


In the Course Schedule problem, we model the courses and their prerequisites as a directed graph. Each course is treated as a node, and a prerequisite relationship [a, b] means there is a directed edge from course a to course b, indicating that a depends on b. The core idea is that we can finish all courses only if the graph does not contain a cycle. If there is a cycle, it means there is a circular dependency — for example, course A requires B, B requires C, and C requires A — which makes it impossible to complete all courses.

To detect cycles, we use Depth-First Search (DFS) along with a 3-state tracking system: UNVISITED, VISITING, and VISITED. When we start exploring a node, we mark it as VISITING. If during DFS we encounter a node that is already marked as VISITING, it means we have found a cycle because we came back to a node that is still in the current recursion path. If we finish exploring all neighbors of a node without finding a cycle, we mark it as VISITED, meaning it is safe.

We run DFS on all courses because the graph may not be fully connected. If any DFS call detects a cycle, we return False. If no cycles are found after checking all nodes, we return True, meaning all courses can be completed. The time complexity is O(V + E), where V is the number of courses and E is the number of prerequisite relationships, since we visit each node and edge once
