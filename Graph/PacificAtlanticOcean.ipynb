1️⃣ Understand the Goal

We need to find all cells from which water can flow to both:
Pacific Ocean (top row + left column)
Atlantic Ocean (bottom row + right column)

Water flows:
From higher → lower or equal height

2️⃣ Key Insight (Reverse Thinking)

Instead of:
Starting from every cell and trying to reach oceans

We do:
Start from the oceans and move inward

Why?
Because if water can flow downhill from A → B,
then in reverse we can move from B → A if height allows.

3️⃣ Build Strategy

Create two sets:

pac → cells reachable from Pacific
atl → cells reachable from Atlantic

Run DFS:

From top row + left column (Pacific)
From bottom row + right column (Atlantic)

Move only to neighbors with:
neighbor_height >= current_height

4️⃣ Final Step

Take intersection:

Cells that are in both pac and atl
Those cells can flow to both oceans.

🔹 Why This Works
Each cell is visited at most once per ocean.
So total time = O(m × n).

```python
from typing import List

class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        if not heights or not heights[0]:
            return []

        ROWS, COLS = len(heights), len(heights[0])
        pac, atl = set(), set()

        def dfs(r: int, c: int, visited: set, prev: int) -> None:
            if (
                r < 0 or c < 0 or r >= ROWS or c >= COLS or
                (r, c) in visited or
                heights[r][c] < prev
            ):
                return

            visited.add((r, c))
            h = heights[r][c]
            dfs(r + 1, c, visited, h)
            dfs(r - 1, c, visited, h)
            dfs(r, c + 1, visited, h)
            dfs(r, c - 1, visited, h)

        # Start DFS from Pacific borders (top row, left col)
        for c in range(COLS):
            dfs(0, c, pac, heights[0][c])
        for r in range(ROWS):
            dfs(r, 0, pac, heights[r][0])

        # Start DFS from Atlantic borders (bottom row, right col)
        for c in range(COLS):
            dfs(ROWS - 1, c, atl, heights[ROWS - 1][c])
        for r in range(ROWS):
            dfs(r, COLS - 1, atl, heights[r][COLS - 1])

        # Cells reachable from both
        res = []
        for r in range(ROWS):
            for c in range(COLS):
                if (r, c) in pac and (r, c) in atl:
                    res.append([r, c])

        return res
```

This problem can be modeled as a graph traversal problem on a grid. Each cell represents a node, and water can flow from a cell to its neighboring cells if the neighbor’s height is less than or equal to the current cell’s height. The goal is to find all cells from which water can flow to both the Pacific and Atlantic oceans.

A brute force approach would be to start DFS or BFS from every cell and check if it can reach both oceans, but that would be inefficient because it would repeat a lot of work. Instead, we reverse the thinking. Rather than starting from every cell and trying to reach the oceans, we start from the ocean borders and move inward.

For the Pacific, we start DFS from the top row and left column. For the Atlantic, we start from the bottom row and right column. While traversing, we only move to neighboring cells whose height is greater than or equal to the current cell’s height. This reverse condition ensures that if water can flow from cell A to cell B in the original direction, then we can move from B to A in our traversal.

We maintain two sets to track cells reachable from the Pacific and Atlantic. Finally, the cells that appear in both sets are the answer. Each cell is visited at most once per ocean, so the time complexity is O(m × n), and space complexity is also O(m × n).
