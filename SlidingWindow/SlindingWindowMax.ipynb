nums = [1,3,-1,-3,5,3,6,7]
k = 3

Windows:

[1,3,-1] → 3
[3,-1,-3] → 3
[-1,-3,5] → 5
[-3,5,3] → 5
[5,3,6] → 6
[3,6,7] → 7

Output:

[3,3,5,5,6,7]

```python
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        res = []  

        # Loop through all possible window starting positions or moves the window
        # The last window starts at index len(nums) - k
        for i in range(len(nums) - k + 1):
            window_max = nums[i]

            # Check all elements inside the current window or checks element inside the window
            for j in range(i, i + k):
                window_max = max(window_max, nums[j])

            res.append(window_max)

        return res

# O(n * k)
# Because for every window we scan k elements.
```

The formula n - k + 1 represents the number of possible starting positions for a window of size k inside an array of length n. The last valid starting index is n - k because the window must still contain k elements. Therefore the total number of windows is (n - k) + 1, which simplifies to n - k + 1.

In the brute force approach, we simulate the sliding window directly. The idea is to move a window of size k across the array and compute the maximum element for each window. For every possible starting position of the window, we examine all k elements inside that window and track the largest value. After finding the maximum, we store it in the result list and move the window one step to the right. We repeat this process until the window reaches the end of the array.

The number of possible windows is n - k + 1, where n is the length of the array. For each window, we scan k elements to find the maximum, which results in a time complexity of O(n × k). The space complexity is O(n - k + 1) for storing the result. While this approach is straightforward and easy to implement, it becomes inefficient when the array size or window size is large, which is why more optimized solutions use a monotonic deque to achieve O(n) time complexity.

```python
import collections
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        output = []
        q = collections.deque()  # stores indices
        l = r = 0

        while r < len(nums):

            # remove smaller values from the right
            while q and nums[q[-1]] < nums[r]:
                q.pop()

            q.append(r)

            # remove left value from window
            if l > q[0]:
                q.popleft()

            # once window size reaches k
            if (r + 1) >= k:
                output.append(nums[q[0]])
                l += 1

            r += 1

        return output 
```

The optimal solution uses a monotonic deque to keep track of useful elements in the current sliding window. The deque stores indices of elements and maintains them in decreasing order of their values, so the largest element is always at the front. As we iterate through the array, we first remove elements from the back of the deque if they are smaller than the current element, because they can never become the maximum in the presence of a larger element. Then we append the current index to the deque. If the index at the front of the deque falls outside the current window, we remove it. Once the window size reaches k, the element at the front of the deque represents the maximum for that window, and we add it to the result. Since each element is inserted and removed from the deque at most once, the time complexity is O(n) and the space complexity is O(k).
