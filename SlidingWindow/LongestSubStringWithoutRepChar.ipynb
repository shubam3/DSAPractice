```python
# Given a string s, find the length of the longest substring without duplicate characters.

# Example 1:

# Input: s = "abcabcbb"
# Output: 3
# Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.


class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        last_index = {}
        left = 0
        longest = 0

        for right,ch in enumerate(s):
            if ch in last_index and last_index[ch] >= left: 
               left = last_index[ch] + 1

            last_index[ch] = right
            longest = max(longest,right-left+1)

        return longest       
```

This solution uses the sliding window technique to find the length of the longest substring without repeating characters. We maintain two pointers: left represents the start of the current window, and right expands the window as we iterate through the string. A dictionary last_index stores the most recent index of each character. If we encounter a character that already exists in the current window (meaning its last seen index is greater than or equal to left), we move the left pointer to one position after that character’s previous index to remove the duplicate from the window. At each step, we calculate the current window length (right - left + 1) and update the maximum length found so far. This approach ensures each character is processed at most twice, giving a time complexity of O(n) and space complexity of O(n) for storing character indices.

```python
def lengthOfLongestSubstring(s: str) -> int:
    n = len(s)
    max_length = 0

    for i in range(n):
        seen = set()
        for j in range(i, n):
            if s[j] in seen:
                break
            seen.add(s[j])
            max_length = max(max_length, j - i + 1)

    return max_length
```

The brute force approach checks all possible substrings and verifies whether each substring contains duplicate characters.

The idea is to use two nested loops: the outer loop selects a starting index, and the inner loop expands the substring one character at a time. For each substring, we use a set to check whether a character has already appeared. If we find a duplicate, we stop expanding that substring and move to the next starting index. While doing this, we keep track of the maximum length found.

This approach is simple to understand but inefficient. The time complexity is O(n²) because we generate all substrings, and checking duplicates adds extra work. The space complexity is O(n) due to the set used for duplicate checking.
