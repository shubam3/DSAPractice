```python
from typing import List

class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Rotates the matrix in-place by 90 degrees clockwise.
        Do not return anything; modify matrix directly.
        """
        n = len(matrix)

        # 1️⃣ Transpose the matrix: matrix[i][j] ↔ matrix[j][i]
        for i in range(n):
            for j in range(i + 1, n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

        # 2️⃣ Reverse each row
        for i in range(n):
            matrix[i].reverse()


# -----------------------
# Example usage
# -----------------------
if __name__ == "__main__":
    mat = [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]
    ]

    Solution().rotate(mat)
    print(mat)  # [[7,4,1],[8,5,2],[9,6,3]]
```
