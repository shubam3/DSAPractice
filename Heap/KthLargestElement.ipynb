```python
# Design a class KthLargest that:

# Takes an integer k and an initial array nums

# Continuously receives numbers using add(val)

# After each insertion, returns the kth largest element among all elements seen so far

import heapq

class KthLargest:

    def __init__(self, k: int, nums):
        self.k = k
        self.minHeap = nums
        heapq.heapify(self.minHeap)

        # Keep only k largest elements
        while len(self.minHeap) > k:
            heapq.heappop(self.minHeap)

    def add(self, val: int) -> int:
        heapq.heappush(self.minHeap, val)

        if len(self.minHeap) > self.k:
            heapq.heappop(self.minHeap)

        return self.minHeap[0]
```

In the Kth Largest Element in a Stream problem, we need to continuously maintain the kth largest element as new numbers are added. Instead of sorting the entire list after every insertion, which would be inefficient, we use a min-heap of size k. The key idea is that we only care about the k largest elements seen so far. So we maintain a min-heap that stores exactly those k largest elements. Whenever a new number is added, we push it into the heap. If the heap size exceeds k, we remove the smallest element. This ensures that the heap always contains the k largest elements, and the smallest element inside the heap represents the kth largest overall. Since heap operations take O(log k) time, each insertion is efficient, and the overall space complexity is O(k).
