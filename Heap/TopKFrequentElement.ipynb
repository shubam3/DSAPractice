```python
from collections import Counter
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        
        # Step 1: Count frequency of each number
        frequency = Counter(nums)
        
        # Step 2: Convert dictionary into list of (num, freq) pairs
        items = list(frequency.items())
        
        # Step 3: Sort items based on frequency (descending)
        items.sort(key=lambda x: x[1], reverse=True)
        
        # Step 4: Collect first k elements
        result = []
        for i in range(k):
            result.append(items[i][0])
        
        return result


```

First, we count the frequency of each number using a hashmap. Then we convert it into a list and sort it based on frequency in descending order. Finally, we collect the first k elements from the sorted list. This approach runs in O(n log n) time due to sorting.

```python
import heapq
from collections import Counter
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = Counter(nums)  # Step 1: Count frequency
        
        minHeap = []

        for num, freq in count.items():
            heapq.heappush(minHeap, (freq, num))
            if len(minHeap) > k:
                heapq.heappop(minHeap)

        return [num for freq, num in minHeap]


# Count frequency using a hashmap.

# Maintain a min-heap of size k.

# Heap stores (frequency, number).

# If heap size exceeds k, remove smallest frequency.

# nums = [1,1,1,2,2,3]
# k = 2
# Frequency:

# 1 → 3
# 2 → 2
# 3 → 1
# Top 2 frequent = [1,2]

# 🔹 Complexity
# Counting → O(n)

# Heap operations → O(n log k)

# Space → O(n)
```

We first count the frequency of each element using a hashmap. Then we use a min-heap of size k to keep track of the k most frequent elements. If the heap size exceeds k, we remove the element with the smallest frequency. In the end, the heap contains the k most frequent elements. This approach runs in O(n log k) time and uses O(n) space.

```python
from collections import Counter
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = Counter(nums)
        
        # Create buckets: index = frequency
        buckets = [[] for _ in range(len(nums) + 1)]

        for num, freq in count.items():
            buckets[freq].append(num)

        result = []

        # Traverse buckets in reverse (high freq first)
        for freq in range(len(buckets) - 1, 0, -1):
            for num in buckets[freq]:
                result.append(num)
                if len(result) == k:
                    return result


# Example:

# nums = [1,1,1,2,2,3]

# Frequency:

# 1 → 3
# 2 → 2
# 3 → 1


# Buckets:

# index 1 → [3]
# index 2 → [2]
# index 3 → [1]


# Traverse from back → pick most frequent first.

# Complexity

# Time: O(n)

# Space: O(n)
```

In the bucket sort approach for Top K Frequent Elements, we take advantage of the fact that the maximum possible frequency of any element is at most n, where n is the length of the array. First, we count the frequency of each element using a hashmap. Then, instead of sorting by frequency, we create an array of buckets where the index represents the frequency, and each bucket stores the numbers that appear that many times. After filling the buckets, we iterate from the highest possible frequency down to the lowest, collecting elements until we have gathered k of them. This avoids sorting altogether and gives a time complexity of O(n), since we only traverse the array and the buckets once. The space complexity is also O(n).
