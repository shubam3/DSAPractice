```python
class Solution:
    def wordBreak(self, s, wordDict):
        
        # Convert list to set for O(1) lookup
        wordSet = set(wordDict)

        # Recursive function
        # start = current index in string
        def dfs(start):

            # Base case:
            # If we reached end of string,
            # it means segmentation is successful
            if start == len(s):
                return True

            # Try every possible end index
            # to create a substring from start -> end
            for end in range(start + 1, len(s) + 1):

                # Check if current substring is in dictionary
                if s[start:end] in wordSet:

                    # Recursively check the remaining part
                    if dfs(end):
                        return True

            # No valid segmentation found
            return False

        # Start recursion from index 0
        return dfs(0)
   

```

In the recursive approach, I try to split the string at every possible position. At each index, I check if the prefix exists in the dictionary, and if it does, I recursively verify whether the remaining substring can also be segmented. If any path reaches the end of the string, the answer is true. However, this approach is inefficient because the same starting indices are recalculated many times, leading to exponential time complexity.

To optimize this, I use memoization. I store the result for each starting index in a hash map, indicating whether the substring from that index can be segmented or not. Before solving a subproblem, I first check the memo to avoid recomputation. This reduces the time complexity significantly to O(n²), since each index is processed once with substring checks, while maintaining O(n) extra space.

```python
class Solution:
    def wordBreak(self, s: str, wordDict) -> bool:
        wordSet = set(wordDict)   # O(1) lookup
        memo = {}                 # start_index -> True/False

        def dfs(start):
            # If we reached the end, we successfully segmented the string
            if start == len(s):
                return True

            # If already computed, reuse result
            if start in memo:
                return memo[start]

            # Try all possible next words starting at 'start'
            for end in range(start + 1, len(s) + 1):
                if s[start:end] in wordSet:
                    if dfs(end):
                        memo[start] = True
                        return True

            # No valid split from this start index
            memo[start] = False
            return False

        return dfs(0)
```

In the dynamic programming approach, I define dp[i] as whether the substring s[0:i] can be segmented into valid dictionary words. I initialize dp[0] = true because an empty string is always segmentable. Then, for each position i, I try all possible split points j before it. If the left part s[0:j] is already segmentable and the right part s[j:i] exists in the dictionary, I mark dp[i] as true. This means the current prefix can be formed from smaller valid subproblems. The final answer is dp[n], where n is the string length. This bottom-up approach avoids repeated recursion, runs in O(n²) time, and uses O(n) space.

```python
class Solution:
    def wordBreak(self, s: str, wordDict) -> bool:
        
        # Convert list to set for fast O(1) lookup
        wordSet = set(wordDict)

        n = len(s)

        # dp[i] = True if substring s[0:i]
        # can be segmented into dictionary words
        dp = [False] * (n + 1)

        # Base case:
        # Empty string can always be segmented
        dp[0] = True

        # Build DP table from left to right
        for i in range(1, n + 1):

            # Try every possible split point
            for j in range(i):

                # Check:
                # 1) left part is valid (dp[j] == True)
                # 2) right part s[j:i] is a dictionary word
                if dp[j] and s[j:i] in wordSet:
                    dp[i] = True
                    break   # no need to check more splits

        # Final answer: can whole string be segmented?
        return dp[n]
```
