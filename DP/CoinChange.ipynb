```python
def coinChange(coins, amount):

    if amount == 0:
        return 0

    if amount < 0:
        return float('inf')

    min_coins = float('inf')

    for coin in coins:
        result = 1 + coinChange(coins, amount - coin)
        min_coins = min(min_coins, result)

    return min_coins if min_coins != float('inf') else -1
```

The Coin Change problem can first be solved using recursion. At each step, I try every coin and recursively compute the minimum coins needed for amount - coin. I take the minimum among all possibilities and add one for the current coin. However, this pure recursive approach is inefficient because it recomputes the same subproblems many times, leading to exponential time complexity.

To optimize this, I use memoization. I store the result for each remaining amount in a dictionary so that each subproblem is solved only once. Before computing a state, I check if it’s already cached. This reduces the time complexity to O(amount × number of coins) and the space complexity to O(amount), making the solution efficient and suitable for larger inputs.

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        memo = {}  # amount -> min coins

        def dfs(rem):
            # Base cases
            if rem == 0:
                return 0
            if rem < 0:
                return float('inf')

            # Return cached answer if already computed
            if rem in memo:
                return memo[rem]

            best = float('inf')

            # Try every coin and take the minimum
            for coin in coins:
                best = min(best, 1 + dfs(rem - coin))

            memo[rem] = best
            return best

        ans = dfs(amount)
        return -1 if ans == float('inf') else ans
```

For the Coin Change problem, I use a bottom-up dynamic programming approach. I define dp[i] as the minimum number of coins required to make amount i. I initialize dp[0] = 0 since zero coins are needed for amount zero, and set all other values to infinity. Then for each amount from 1 to the target, I iterate through all available coins. If a coin can contribute to the current amount, I update dp[i] as the minimum of its current value and 1 + dp[i - coin]. This builds the solution iteratively using previously computed smaller amounts. The final answer is stored in dp[amount]. This approach runs in O(amount × number of coins) time and uses O(amount) space, which is optimal for this problem.

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        # Initialize DP array
        dp = [float('inf')] * (amount + 1)
        
        # Base case
        dp[0] = 0   # 0 coins needed for amount 0

        # Build solution from 1 to amount
        for i in range(1, amount + 1):
            for coin in coins:
                if i - coin >= 0:
                    dp[i] = min(dp[i], 1 + dp[i - coin])

        # If still infinity, return -1
        return dp[amount] if dp[amount] != float('inf') else -1
```

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        
        # Create DP array of size (amount + 1)
        # dp[i] will store the minimum number of coins needed to make amount i
        dp = [float('inf')] * (amount + 1)

        # Base case:
        # 0 coins are needed to make amount 0
        dp[0] = 0

        # Build the solution from 1 up to the target amount
        for i in range(1, amount + 1):

            # Try every coin
            for coin in coins:

                # Check if the coin can be used
                # (i - coin must not go negative)
                if i - coin >= 0:

                    # If we use this coin:
                    # 1 coin (current) + best solution for remaining amount (i - coin)
                    dp[i] = min(dp[i], 1 + dp[i - coin])

        # If dp[amount] is still infinity,
        # it means the amount cannot be formed
        if dp[amount] == float('inf'):
            return -1

        # Otherwise return the minimum coins required
        return dp[amount]
```
