```python
class Solution:
    def numDecodings(self, s: str) -> int:
        # Length of string
        n = len(s)

        # Edge case: empty string
        if n == 0:
            return 0

        # dp1 represents dp[i+1]
        # Base case: empty substring has 1 way to decode
        dp1 = 1

        # dp2 represents dp[i+2]
        # Initially just a placeholder (will shift properly)
        dp2 = 0

        # Traverse string from right to left
        for i in range(n - 1, -1, -1):

            # If current character is '0',
            # it cannot be decoded alone
            if s[i] == "0":
                curr = 0

            else:
                # Take single digit decode
                # Ways = dp[i+1]
                curr = dp1

                # Check if two-digit decode is valid (10 to 26)
                if (
                    i + 1 < n and
                    (s[i] == "1" or
                     (s[i] == "2" and s[i + 1] in "0123456"))
                ):
                    # Add ways from dp[i+2]
                    curr += dp2

            # Shift the DP window for next iteration
            dp2 = dp1
            dp1 = curr

        # dp1 now holds dp[0]
        return dp1
```

Why do we loop from the last character?

Because the recurrence relation depends on future indices:

dp[i]=dp[i+1]+dp[i+2]

That means:

To compute dp[i]
We need values of dp[i+1] and dp[i+2]

So those must already be calculated.


What does dp[i] mean?
dp[i] = number of ways to decode substring s[i:]

So we build answers from the end toward the beginning.

We iterate from right to left because each state depends on future states (i+1 and i+2), so they must be computed first.

The Decode Ways problem can be solved using dynamic programming. I define dp[i] as the number of ways to decode the substring starting at index i. Since each position depends on the next one or two positions, the recurrence is: if the current digit is non-zero, we can take one digit and add dp[i+1]. If the two-digit number formed with the next character is between 10 and 26, we can also take two digits and add dp[i+2]. To optimize space, instead of storing a full DP array, I keep only two variables representing dp[i+1] and dp[i+2]. I iterate from right to left because each state depends on future states. This approach runs in O(n) time and O(1) space
