```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        
        def dfs(row, col):
            # If out of grid → invalid path
            if row >= m or col >= n:
                return 0

            # If reached bottom-right → found one valid path
            if row == m - 1 and col == n - 1:
                return 1

            # Move down and move right
            return dfs(row + 1, col) + dfs(row, col + 1)

        return dfs(0, 0)
```

Brute Force Approach

In the brute-force approach, I use recursion to explore all possible paths from the top-left to the bottom-right cell. At each position, I make two choices: move right or move down. If I reach the destination, I return 1 to count that path. If I go out of bounds, I return 0. The total paths are the sum of both recursive calls. However, this approach has exponential time complexity because the same subproblems are recomputed multiple times.

🔹 Memoization (Top-Down DP)

To optimize, I store the result for each cell (row, col) in a memo dictionary. Before computing a state, I check if it has already been solved. This prevents recomputation of overlapping subproblems, reducing the time complexity to O(m × n) and space complexity to O(m × n).

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        
        memo = {}

        def dfs(row, col):
            # If out of bounds → no path
            if row >= m or col >= n:
                return 0

            # If reached destination → 1 valid path
            if row == m - 1 and col == n - 1:
                return 1

            # If already computed, return stored value
            if (row, col) in memo:
                return memo[(row, col)]

            # Compute and store result
            memo[(row, col)] = dfs(row + 1, col) + dfs(row, col + 1)

            return memo[(row, col)]

        return dfs(0, 0)
```

1️⃣ Define DP State

Let dp[i][j]=number of ways to reach cell (i,j)


2️⃣ Identify Base Cases

First row → only move right → 1 way
First column → only move down → 1 way


3️⃣ Identify Transition (Recurrence Relation)

For every other cell:

dp[i][j]=dp[i−1][j]+dp[i][j−1]

Reason:
You can only arrive from:

Top cell
Left cell


4️⃣ Build the Table

Start filling from top-left
Move row by row
Use previously computed values

5️⃣ Final Answer

Return:
dp[m−1][n−1]

6️⃣ Complexity

Time: 
O(m×n)

Space: 
O(m×n)

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        
        # Create DP table of size m x n
        dp = [[0] * n for _ in range(m)]

        # First row → only 1 way (move right)
        for col in range(n):
            dp[0][col] = 1

        # First column → only 1 way (move down)
        for row in range(m):
            dp[row][0] = 1

        # Fill the rest of the table
        for row in range(1, m):
            for col in range(1, n):
                dp[row][col] = dp[row - 1][col] + dp[row][col - 1]

        # Bottom-right cell contains answer
        return dp[m - 1][n - 1]
```

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        
        # Create a 1D DP array of size n
        # Initially, first row has only 1 way to reach each cell
        dp = [1] * n

        # Start from second row
        for row in range(1, m):
            
            # Update columns from left to right
            for col in range(1, n):
                
                # dp[col] → value from top (previous row)
                # dp[col - 1] → value from left (current row)
                dp[col] = dp[col] + dp[col - 1]

        # Last element contains total paths
        return dp[n - 1]
```
