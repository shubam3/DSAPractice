The brute-force approach generates all possible substrings using two loops and checks each one to see if it’s a palindrome by comparing it with its reverse. Every palindrome found is counted. This method is easy to understand but inefficient, with O(n³) time complexity because there are O(n²) substrings and each palindrome check takes O(n) time.

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        count = 0

        for i in range(len(s)):
            for j in range(i, len(s)):
                sub = s[i:j+1]
                if sub == sub[::-1]:  # palindrome check
                    count += 1

        return count
```

The optimal approach uses the expand-around-center technique. The key idea is that every palindrome expands symmetrically from its center. For each index, I expand around a single character for odd-length palindromes and around the gap between two characters for even-length palindromes. Each successful expansion counts as one palindromic substring. This approach avoids redundant checks, runs in O(n²) time, and uses O(1) extra space.

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        res = 0
        for i in range(len(s)):
          res += self.countPali(s,i,i)
          res += self.countPali(s,i,i+1)
        return res


    def countPali(self,s,l,r):
        res = 0
        while l >=0 and r<len(s) and s[l]==s[r]:
            res += 1
            l -= 1
            r += 1         

        return res    
```

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        count = 0

        def expand(l, r):
            nonlocal count
            while l >= 0 and r < len(s) and s[l] == s[r]:
                count += 1
                l -= 1
                r += 1

        for i in range(len(s)):
            expand(i, i)     # odd length
            expand(i, i + 1) # even length

        return count
```
