House Robber II is a variation where houses are arranged in a circle, so the first and last houses are adjacent and cannot both be robbed. To handle this circular constraint, I break the problem into two linear subproblems. In the first case, I consider houses from index 0 to n-2, excluding the last house. In the second case, I consider houses from index 1 to n-1, excluding the first house. For each case, I apply the standard House Robber logic where at each house I choose to either rob it and skip the next one or skip it. The final answer is the maximum of these two cases. This approach ensures the circular constraint is respected and can later be optimized using memoization or dynamic programming.

```python
def rob(nums):
    n = len(nums)
    if n == 1:
        return nums[0]

    def rob_from(i, end):
        if i > end:
            return 0

        rob_current = nums[i] + rob_from(i + 2, end)
        skip_current = rob_from(i + 1, end)

        return max(rob_current, skip_current)

    # Case 1: exclude last house
    case1 = rob_from(0, n - 2)

    # Case 2: exclude first house
    case2 = rob_from(1, n - 1)

    return max(case1, case2)


# Example
nums = [2, 3, 2]
print(rob(nums))  # Output: 3
```

In House Robber II, the houses are arranged in a circle, so the first and last houses are adjacent and cannot both be robbed. To handle this circular constraint, I split the problem into two linear subproblems. In the first case, I consider houses from index 0 to n−2, excluding the last house. In the second case, I consider houses from index 1 to n−1, excluding the first house. For each case, I apply the standard bottom-up House Robber dynamic programming approach, where dp[i] represents the maximum money that can be robbed from the first i houses. The transition is choosing the maximum between skipping the current house or robbing it and adding the value from two houses before. Finally, I return the maximum result from the two cases. This approach runs in O(n) time and uses O(n) space, and it can be further optimized to constant space

```python
def rob(nums):
    n = len(nums)
    if n == 1:
        return nums[0]

    def rob_linear(arr):
        m = len(arr)
        dp = [0] * (m + 1)
        dp[0] = 0
        dp[1] = arr[0]

        for i in range(2, m + 1):
            dp[i] = max(dp[i - 1], arr[i - 1] + dp[i - 2])

        return dp[m]

    # Case 1: exclude last house
    case1 = rob_linear(nums[:-1])

    # Case 2: exclude first house
    case2 = rob_linear(nums[1:])

    return max(case1, case2)


# Example
nums = [2, 3, 2]
print(rob(nums))  # 3
```

In House Robber II, houses are arranged in a circle, so the first and last houses cannot both be robbed. To handle this, I split the problem into two linear cases: robbing houses from index 0 to n-2, and from index 1 to n-1. For each case, I apply the space-optimized House Robber dynamic programming approach. Instead of using a DP array, I keep only two variables representing the maximum money robbed up to the previous house and the house before that. For each house, I choose the maximum between skipping it or robbing it and adding its value to the amount from two houses back. I take the maximum result from the two cases. This approach runs in O(n) time and uses O(1) extra space, making it efficient and interview-friendly.

```python
def rob(nums):
    n = len(nums)
    if n == 1:
        return nums[0]

    def rob_linear(start, end):
        prev2 = 0   # dp[i-2]
        prev1 = 0   # dp[i-1]

        for i in range(start, end + 1):
            curr = max(prev1, nums[i] + prev2)
            prev2 = prev1
            prev1 = curr

        return prev1

    case1 = rob_linear(0, n - 2)  # exclude last
    case2 = rob_linear(1, n - 1)  # exclude first

    return max(case1, case2)


# Example
nums = [2, 3, 2]
print(rob(nums))  # 3
```
