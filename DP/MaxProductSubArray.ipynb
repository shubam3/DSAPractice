```python
class Solution:
    def maxProduct(self, nums):
        n = len(nums)
        max_product = float('-inf')

        for i in range(n):
            product = 1
            for j in range(i, n):
                product *= nums[j]
                max_product = max(max_product, product)

        return max_product
```

Brute Force Approach

In the brute-force approach, I generate all possible contiguous subarrays using two nested loops and compute the product of each subarray. I track the maximum product encountered. This approach works but has O(n²) time complexity since we examine all subarrays, and it is inefficient for large inputs.

🔹 Optimal Approach (Dynamic Programming)

The optimal solution runs in O(n) time using dynamic programming. At each index, I track both the maximum and minimum product ending at that position. The minimum is important because multiplying two negative numbers can produce a large positive number. If the current number is negative, I swap the max and min before updating. At each step, I choose whether to start a new subarray from the current number or extend the previous one. I update the global maximum during iteration. This approach uses O(1) space and efficiently handles negative numbers and zeros.

```python
class Solution:
    def maxProduct(self, nums):
        
        # Initialize max, min, and answer with first element
        # max_here → maximum product ending at current index
        # min_here → minimum product ending at current index
        # ans → global maximum product found so far
        max_here = nums[0]
        min_here = nums[0]
        ans = nums[0]

        # Start from second element
        for i in range(1, len(nums)):
            x = nums[i]

            # If current number is negative,
            # swap max and min because sign will flip
            if x < 0:
                max_here, min_here = min_here, max_here

            # Either start new subarray at x,
            # or extend previous max product
            max_here = max(x, x * max_here)

            # Either start new subarray at x,
            # or extend previous min product
            min_here = min(x, x * min_here)

            # Update global maximum
            ans = max(ans, max_here)

        # Return the maximum product subarray
        return ans
```
