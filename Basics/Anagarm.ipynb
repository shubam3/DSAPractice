```python
def anagrams(s1, s2):
   return char_count(s1) == char_count(s2)


def char_count(s):
   count = {}

   for char in s:
     if char not in count:
       count[char] = 0

     count[char] +=1

   return count 
```

This solution checks whether two strings are anagrams by comparing their character frequency counts. The char_count function creates a dictionary where each character in the string is stored as a key and its frequency as the value. It iterates through the string and increments the count for each character. In the anagrams function, we simply compare the frequency dictionaries of both strings. If the dictionaries are equal, it means both strings contain the same characters with the same frequencies, so they are anagrams. This approach runs in O(n) time since we traverse each string once, and the space complexity is O(n) due to the dictionary storing character counts.
