```python
from collections import Counter
def anagrams(s1,s2):
  return Counter(s1) == Counter(s2)
```

Time Complexity
1. char_count(s)

Loop through each character → O(n)

Checking char not in count → O(1) average (hashmap)

Updating the count → O(1)

So:

➝ char_count(s) = O(n)

2. Two calls:

char_count(s1) → O(n)

char_count(s2) → O(m)

3. Dictionary comparison:

Comparing two dicts by keys and values → O(k)
(k = number of unique characters, usually small)
→ O(k) ≈ O(1) in practice for ASCII.

⭐ Final Time Complexity
O(n + m)

If both strings same length:

→ O(n)
✅ Space Complexity

Dictionary stores counts for each unique char.

Worst case: all characters unique → size = n.

Space: O(k) → O(n) worst-case
⭐ Final Answer

Time Complexity: O(n + m)
Space Complexity: O(n)
