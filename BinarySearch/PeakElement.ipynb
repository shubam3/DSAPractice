```python
# A peak element is an element that is strictly greater than its neighbors.

# Given a 0-indexed integer array nums, find a peak element, and return its index. If the array contains multiple peaks, return the index to any of the peaks.

# You may imagine that nums[-1] = nums[n] = -∞. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.

# You must write an algorithm that runs in O(log n) time.

# Example 1:

# Input: nums = [1,2,3,1]
# Output: 2
# Explanation: 3 is a peak element and your function should return the index number 2.
```

```python
from typing import List

class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        n = len(nums)

        for i in range(n):
            left = nums[i - 1] if i > 0 else float('-inf')
            right = nums[i + 1] if i < n - 1 else float('-inf')

            if nums[i] > left and nums[i] > right:
                return i

# For each index:
# Compare with left neighbor
# Compare with right neighbor
# If greater than both → it's a peak

# Return the index
```

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        # Initialize binary search boundaries
        left, right = 0, len(nums) - 1

        # Continue searching while there is more than one element
        while left < right:
            # Find the middle index
            mid = (left + right) // 2

            # Compare mid element with the next element
            if nums[mid] < nums[mid + 1]:
                # If the right neighbor is greater,
                # a peak must exist on the right side
                left = mid + 1
            else:
                # Otherwise, a peak exists on the left side
                # (including mid itself)
                right = mid

        # When left == right, we found the peak index
        return left
```

This solution uses binary search to find a peak element in O(log n) time. The key insight is that if the middle element is smaller than its right neighbor, then a peak must exist on the right side because the slope is increasing. Otherwise, if the middle element is greater than or equal to its right neighbor, then a peak must exist on the left side, including the middle itself. Based on this comparison, we eliminate half of the search space in each step. We continue this process until left equals right, which will point to a peak element. The time complexity is O(log n) due to binary search, and the space complexity is O(1).
