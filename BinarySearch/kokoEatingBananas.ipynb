```python
import math
from typing import List

class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        maxPile = max(piles)

        for k in range(1, maxPile + 1):
            hours = 0
            for p in piles:
                hours += math.ceil(p / k)

            if hours <= h:
                return k

# Time: O(maxPile * n) (slow if piles are large)
# Space: O(1)
```

The brute force approach for Koko Eating Bananas works by trying every possible eating speed from 1 up to the maximum pile size. For each speed k, we calculate the total number of hours required to finish all piles using the formula ceil(pile / k) for each pile. We sum these hours and check whether the total is less than or equal to h. The first speed that satisfies the condition is our answer because we are testing speeds in increasing order. However, this approach is inefficient since it may check many possible speeds, giving a time complexity of O(maxPile × n), where n is the number of piles.

```python
import math
from typing import List

class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        left, right = 1, max(piles)

        def can_finish(k: int) -> bool:
            hours = 0
            for p in piles:
                hours += (p + k - 1) // k   # ceil(p/k) without float
            return hours <= h

        while left < right:
            mid = (left + right) // 2
            if can_finish(mid):
                right = mid   # mid works, try smaller
            else:
                left = mid + 1  # mid too slow, go bigger

        return left


# Time: O(n log maxPile)
# Space: O(1)

# Key idea

# Minimum speed = 1

# Maximum speed = max(piles)

# If a speed k finishes within h hours, try smaller speed.

# Otherwise, need bigger speed.
```

The optimal solution for Koko Eating Bananas uses binary search on the possible eating speed instead of checking every speed one by one. The minimum possible speed is 1 and the maximum possible speed is the largest pile. For any chosen speed k, we calculate the total hours required using the formula ceil(pile / k) for each pile. If the total hours are less than or equal to h, it means this speed works, so we try to find a smaller valid speed on the left side. If it exceeds h, the speed is too slow, so we search on the right side. Since we eliminate half of the search space in each step, the time complexity becomes O(n log maxPile), and the space complexity is O(1).
