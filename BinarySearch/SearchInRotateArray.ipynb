```python
# ⏱ Time Complexity:
# O(n) – you might have to check every element

# 🧠 Space Complexity:
# O(1) – no extra space used

def search(nums, target):
    # Step 1: Go through each element
    for i in range(len(nums)):
        # Step 2: Check if it's equal to the target
        if nums[i] == target:
            return i  # Found the target, return its index
    return -1  # Not found
```

```python
# ⏱ Time Complexity:
# O(log n) — standard binary search logic

# 🧠 Space Complexity:
# O(1) — constant space


def search(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        # Step 1: Found the target
        if nums[mid] == target:
            return mid

        # Step 2: Determine which half is sorted
        if nums[left] <= nums[mid]:
            # Left half is sorted
            if nums[left] <= target < nums[mid]:
                right = mid - 1  # target is in the left half
            else:
                left = mid + 1   # target is in the right half
        else:
            # Right half is sorted
            if nums[mid] < target <= nums[right]:
                left = mid + 1   # target is in the right half
            else:
                right = mid - 1  # target is in the left half

    return -1  # Target not found
```

This problem searches a target in a rotated sorted array. A normal binary search cannot be applied directly because the array is rotated, but at least one half of the array is always sorted.

So in every iteration:

I calculate the mid index.

If nums[mid] equals the target, I return the index.

Otherwise, I check which half is sorted:

If nums[left] <= nums[mid], the left half is sorted.

Otherwise, the right half is sorted.

Once I know which half is sorted, I check if the target lies inside that sorted range:

If yes, I search in that half.

Otherwise, I search in the other half.

This way I reduce the search space by half each time, similar to binary search.

Time Complexity: O(log n)
Space Complexity: O(1).

Key idea: Even though the array is rotated, one side is always sorted, which helps us decide where to continue the search.
