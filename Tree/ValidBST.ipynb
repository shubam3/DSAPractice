```python
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
       def isBST(root,low,high):
          if not root:
             return True

          if not (low < root.val < high):
            return False

          return isBST(root.left,low,root.val) and isBST(root.right,root.val,high)

       return isBST(root, float('-inf'), float('inf'))   
```

This solution validates a Binary Search Tree using the range constraint approach. The key idea is that every node in a BST must lie within a valid range, not just compared to its immediate parent. We recursively pass down a lower and upper bound for each node. Initially, the root can take any value between negative infinity and positive infinity. For the left child, we update the upper bound to the current node’s value, and for the right child, we update the lower bound to the current node’s value. If any node violates its allowed range, we return False. This ensures the BST property holds globally across the tree. The time complexity is O(n) since we visit each node once, and the space complexity is O(h) due to recursion stack.

```python
class Solution:
    def isValidBST(self, root):
        self.prev = float('-inf')

        def inorder(node):
            if not node:
                return True
            
            if not inorder(node.left):
                return False
            
            # Check BST property
            if node.val <= self.prev:
                return False
            
            self.prev = node.val
            
            return inorder(node.right)

        return inorder(root)
```

The inorder approach works because in a valid Binary Search Tree, performing an inorder traversal (Left → Root → Right) produces values in strictly increasing order. So while traversing the tree, we keep track of the previously visited node’s value. For each current node, we check whether its value is greater than the previous value. If it is not strictly greater, it means the BST property is violated and we return False. If the entire traversal completes without any violations, the tree is valid. This method works because inorder traversal naturally reveals the sorted order property of a BST. The time complexity is O(n) since every node is visited once, and the space complexity is O(h) due to the recursion stack.
