| Traversal     | Order                   | Theory (simple)                  |
| ------------- | ----------------------- | -------------------------------- |
| **Preorder**  | **Root → Left → Right** | Visit the node *before* children |
| **Inorder**   | **Left → Root → Right** | Left nodes first, then root      |
| **Postorder** | **Left → Right → Root** | Children first, root at the end  |

Preorder
✔ Process root first
✔ Then go left
✔ Then go right

Inorder
✔ Visit entire left subtree
✔ Then root
✔ Then right

Postorder
✔ Go deep left
✔ Then deep right
✔ Process root at last

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def preorder(root):
    if root is None:
        return
    print(root.val, end=" ")   # root
    preorder(root.left)        # left
    preorder(root.right)       # right


def inorder(root):
    if root is None:
        return
    inorder(root.left)         # left
    print(root.val, end=" ")   # root
    inorder(root.right)        # right


def postorder(root):
    if root is None:
        return
    postorder(root.left)       # left
    postorder(root.right)      # right
    print(root.val, end=" ")   # root

# Tree:
#      1
#     / \
#    2   3

root = Node(1)
root.left = Node(2)
root.right = Node(3)

print("Preorder: ", end=""); preorder(root)
print("\nInorder: ", end=""); inorder(root)
print("\nPostorder:", end=""); postorder(root)
```

1. Inorder (Left → Root → Right)

In a Binary Search Tree (BST), inorder traversal gives sorted order.

Useful when you need elements in increasing order.

Time Complexity: O(n)

Space Complexity: O(h) (recursion stack)

👉 Root is visited in the middle.

🔹 2. Preorder (Root → Left → Right)

Used to copy or clone a tree.

Useful for serialization of a tree.

Helps build tree structure from traversal.

Root is processed before children.

👉 Root comes first.

🔹 3. Postorder (Left → Right → Root)

Used when you must process children before parent.

Common in deleting/freeing a tree.

Useful in problems like expression tree evaluation.

👉 Root comes last.

```python
from collections import deque

class Solution:
    def levelOrder(self, root):
        if not root:
            return []

        result = []
        queue = deque([root])

        while queue:
            level = []
            size = len(queue)

            for _ in range(size):
                node = queue.popleft()
                level.append(node.val)

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)

            result.append(level)

        return result



# How it works

# Use a queue (FIFO)

# Process nodes level by level

# Add left and right children to queue

# Store values per level

# ⏱ Complexity

# Time: O(n)

# Space: O(n)
```

This approach uses Breadth-First Search (BFS) to traverse a binary tree level by level. We use a queue data structure because BFS follows a First-In-First-Out pattern. We start by adding the root node to the queue. Then, while the queue is not empty, we process all nodes at the current level by looping through the queue’s current size. For each node, we record its value and add its left and right children to the queue if they exist. This ensures we visit nodes level by level from top to bottom and left to right. The time complexity is O(n) because we visit each node once, and the space complexity is O(n) due to the queue storing nodes.
