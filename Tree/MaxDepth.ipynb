```python
class Solution:
    def maxDepth(self, root):
        if not root:
            return 0
        
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        
        return 1 + max(left_depth, right_depth)


# Idea

# Depth = 1 (current node) + max(depth of left, depth of right)

# Time: O(n)
# Space: O(h) (recursion stack)
```

n the recursive DFS approach, we compute the maximum depth by exploring the tree down to its leaf nodes. For each node, we recursively calculate the depth of its left and right subtrees. The depth of the current node is 1 plus the maximum of those two depths. The base case occurs when the node is null, where we return 0. This works because the height of a tree is defined as the longest path from the root to a leaf. The time complexity is O(n) since we visit each node once, and the space complexity is O(h), where h is the height of the tree due to the recursion stack.

```python
from collections import deque

class Solution:
    def maxDepth(self, root):
        if not root:
            return 0
        
        queue = deque([root])
        depth = 0
        
        while queue:
            for _ in range(len(queue)):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            depth += 1
        
        return depth

# Time: O(n)
# Space: O(n)
```

In the BFS approach, we traverse the tree level by level using a queue. We start by adding the root to the queue. For each level, we process all nodes currently in the queue and then add their children. After finishing one full level, we increment the depth counter. When the queue becomes empty, the depth counter represents the maximum depth of the tree. This approach also runs in O(n) time because each node is visited once, and the space complexity is O(n) in the worst case when the tree is wide.
