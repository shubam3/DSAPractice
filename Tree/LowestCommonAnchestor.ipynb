```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if p.val < root.val and q.val < root.val:
            return self.lowestCommonAncestor(root.left,p,q)
        elif p.val > root.val and q.val > root.val:
            return self.lowestCommonAncestor(root.right,p,q)   
        else: 
            return root
```

This solution finds the Lowest Common Ancestor (LCA) in a Binary Search Tree using the BST property. In a BST, all values in the left subtree are smaller than the root, and all values in the right subtree are greater. So, if both p and q are smaller than the current root, the LCA must lie in the left subtree, and we recursively move left. If both are greater than the root, we move right. If one value is smaller and the other is larger (or one equals the root), it means the current root is the split point where the paths to p and q diverge, so it is the LCA. This approach runs in O(h) time, where h is the height of the tree.

```python
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        while root:
            if p.val < root.val and q.val < root.val:
                root = root.left
            elif p.val > root.val and q.val > root.val:
                root = root.right
            else:
                return root

# Instead of recursion, we use a while loop.

# We keep moving left or right until we find the split point.

# When one value is on each side (or equal to root), that node is the LCA.

# Time: O(h) (height of tree)

# Space: O(1) (no recursion stack)
```

The iterative solution for Lowest Common Ancestor in a Binary Search Tree uses the BST property to navigate down the tree without recursion. Starting from the root, we compare the values of p and q with the current node. If both values are smaller than the current node, the LCA must be in the left subtree, so we move left. If both values are greater, we move right. If one value is smaller and the other is larger (or one equals the current node), it means we have found the split point where the paths to p and q diverge, so the current node is the LCA. This approach runs in O(h) time, where h is the height of the tree, and uses O(1) space since it does not rely on recursion.
