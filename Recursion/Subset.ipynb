```python
# Given an integer array nums of unique elements, return all possible subsets (the power set).

# The solution set must not contain duplicate subsets. Return the solution in any order.

 

# Example 1:

# Input: nums = [1,2,3]
# Output: [[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]



def subsets(nums):
    result = []

    def backtrack(index, current):
        if index == len(nums):
            result.append(current[:])
            return

        # Include the element
        current.append(nums[index])
        backtrack(index + 1, current)

        # Exclude the element
        current.pop()
        backtrack(index + 1, current)

    backtrack(0, [])
    return result
```

This brute force solution generates all possible subsets using recursion and backtracking. The key idea is that for each element in the array, we have two choices: either include it in the current subset or exclude it. By recursively making these two decisions at every index, we explore all possible combinations. When we reach the end of the array, we add the current subset to the result. Since each element has two choices, the total number of subsets generated is 2ⁿ. Therefore, the time complexity is O(2ⁿ), and the space complexity is also O(2ⁿ) due to storing all subsets.
