```python
# Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

# Implement the MinStack class:

# MinStack() initializes the stack object.
# void push(int val) pushes the element val onto the stack.
# void pop() removes the element on the top of the stack.
# int top() gets the top element of the stack.
# int getMin() retrieves the minimum element in the stack.
# You must implement a solution with O(1) time complexity for each function.
```

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        
        # Push min value so far
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        if self.stack:
            if self.stack[-1] == self.min_stack[-1]:
                self.min_stack.pop()
            self.stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.min_stack[-1]
```

A MinStack is a stack that supports normal stack operations (push, pop, top) and also returns the minimum element in O(1) time.

To achieve this, we use two stacks:

1️⃣ stack → stores all elements
2️⃣ min_stack → stores the minimum value at each stage

Working Idea

When pushing a value, we also push it into min_stack if it is smaller or equal to the current minimum.

When popping, if the popped value equals the top of min_stack, we remove it from min_stack as well.

The top of min_stack always contains the current minimum.

Time Complexity
Operation	Time
push	O(1)
pop	O(1)
top	O(1)
getMin	O(1)
One-Line Interview Explanation

MinStack uses two stacks: one for normal elements and another to track the minimum element at each stage, allowing getMin() to run in constant time.
