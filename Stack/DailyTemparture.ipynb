```python
from typing import List

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        n = len(temperatures)
        res = [0] * n

        for i in range(n):
            for j in range(i + 1, n):
                if temperatures[j] > temperatures[i]:
                    res[i] = j - i
                    break  # first warmer day found

        return res
    

# Input: temperatures = [73,74,75,71,69,72,76,73]
# Output: [1,1,4,2,1,1,0,0]    
```

This brute force approach solves the Daily Temperatures problem by checking each day individually and searching for the next warmer day. For every index i, we iterate forward from i + 1 to the end of the array and compare temperatures. As soon as we find a day j where the temperature is higher than the current day, we store the difference j - i as the number of days to wait and break the inner loop. If no warmer day is found, we leave the value as 0. Since we use two nested loops, the time complexity is O(n²), and the space complexity is O(n) for storing the result array.

```python
# We use a monotonic decreasing stack to store indices of temperatures.
# The stack keeps temperatures in decreasing order.

# When we find a warmer temperature, we pop from the stack and calculate how many days it took.

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        
        n = len(temperatures)

        # Result array initialized with 0
        # res[i] will store number of days to wait for a warmer temperature
        res = [0] * n

        # Stack will store indices of temperatures
        # Temperatures in stack will be in decreasing order
        stack = []

        # Traverse all temperatures
        for i, t in enumerate(temperatures):

            # If current temperature is greater than the temperature
            # at the index stored on top of the stack
            while stack and t > temperatures[stack[-1]]:

                # Pop the previous day's index
                prev = stack.pop()

                # Calculate number of days waited for a warmer temperature
                res[prev] = i - prev

            # Push current index into stack
            # This day is waiting for a future warmer temperature
            stack.append(i)

        return res


# Time: O(n) (each index pushed/popped once)
# Space: O(n) (stack + result)

# 1️⃣ Initialize variables
# 2️⃣ Loop through temperatures
# 3️⃣ Check warmer temperature : If current temperature is warmer than previous temperatures in stack.
# 4️⃣ Pop and update result
# 5️⃣ Push current index : We store the index for future comparison.
```

This solution uses a monotonic decreasing stack to solve the Daily Temperatures problem in optimal time. The idea is to store the indices of days whose warmer temperature has not yet been found. As we iterate through the array, if the current temperature is higher than the temperature at the index on top of the stack, it means we have found the next warmer day for that index. We then pop that index from the stack and calculate the number of days waited as the difference between the current index and the popped index. We continue this process until the stack is empty or the current temperature is not warmer. Then we push the current index onto the stack. Since each index is pushed and popped at most once, the time complexity is O(n), and the space complexity is O(n).

A monotonic stack is a stack where elements are kept in a specific order:

Monotonic Increasing Stack → elements increase from bottom to top

Monotonic Decreasing Stack → elements decrease from bottom to top

We remove elements from the stack if the order breaks.
