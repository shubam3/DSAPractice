```python
class Solution:
    def isValid(self, s: str) -> bool:
      stack = []
      pairs = {')': '(', '}': '{', ']': '['}  # Mapping of closing to opening brackets

      for char in s:
        if char in pairs:  # If it's a closing bracket
            if not stack or stack.pop() != pairs[char]:  
                return False  # Stack is empty or brackets don't match
        else:
            stack.append(char)  # Push opening brackets

      return not stack  # Return True if stack is empty (all brackets matched)
    

# 1️⃣ Create stack and bracket map    
# 2️⃣ Traverse the string
# 3️⃣ If it is a closing bracket Case 1: stack empty , Case 2: bracket mismatch
# 4️⃣ If it is an opening bracket
# 5️⃣ Final check
```

This solution checks whether a string of brackets is valid using a stack. The idea is that every closing bracket must match the most recent unmatched opening bracket. I use a stack to store opening brackets as I iterate through the string. If I encounter an opening bracket, I push it onto the stack. If I encounter a closing bracket, I check whether the stack is empty or whether the top of the stack matches the corresponding opening bracket using the mapping dictionary. If it doesn’t match, the string is invalid. After processing the entire string, if the stack is empty, it means all brackets were properly matched. The time complexity is O(n), and the space complexity is O(n) in the worst case.
