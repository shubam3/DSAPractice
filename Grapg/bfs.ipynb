```python
from collections import deque

class Solution:
    # Function to return Breadth First Search Traversal of given graph.
    def bfs(self, adj):
        # code here
        
        queue = deque()
        ans = []
        visted = set()
        
        queue.append(0)
        visted.add(0)
        
        while queue:
            e = queue.popleft()
            ans.append(e)
            
            for node in adj[e]:
                if node not in visted:
                    queue.append(node)
                    visted.add(node)
                    
        return ans            
```

```python
from collections import deque

def bfs(graph, start):
    """
    Perform BFS on a graph from a given start node.

    graph : dict
        Adjacency list representation of the graph.
        Example: { 'A': ['B','C'], 'B': ['D'], 'C': ['E'], 'D': [], 'E': [] }
    start : str
        Starting node for BFS.

    Returns:
        visited_order : list
            The order in which nodes are visited.
    """

    visited = set()        # to keep track of visited nodes
    queue = deque([start]) # queue for BFS (FIFO order)
    order = []             # store BFS traversal order

    while queue:
        node = queue.popleft()  # take element from front of queue
        if node not in visited:
            visited.add(node)   # mark node as visited
            order.append(node)  # record the visit

            # enqueue all unvisited neighbors
            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    queue.append(neighbor)

    return order


# -------- Example Usage --------
if __name__ == "__main__":
    # Graph represented as adjacency list
    graph = {
        'A': ['B', 'C'],
        'B': ['D', 'E'],
        'C': ['F'],
        'D': [],
        'E': ['F'],
        'F': []
    }

    print("BFS traversal starting from A:", bfs(graph, 'A'))
```
