1. Build the adjacency list

For each directed edge (u → v), add v to adj[u].
→ This represents that u must come before v.

2. Initialize helper structures

visited[V] → tracks whether a node is already explored.

stack → stores nodes in reverse topological order.

3. Perform DFS on each unvisited node

For every vertex i from 0 to V-1:

If visited[i] == 0, call dfs(i).

4. DFS traversal logic

Inside dfs(node):

Mark the node as visited.

For every adjacent node adjNode of node:

If not visited, recursively call dfs(adjNode).

After exploring all adjacents, push node into the stack.

5. Reverse the stack

Once all nodes are processed, reverse the stack → gives the Topological Order.

```python
# topological_sort.py

class Solution:
    
    def dfs(self, current_node, visited, stack, adj_list):
        visited[current_node] = 1
        for adjNode in adj_list[current_node]:
            if visited[adjNode] == 0:
                self.dfs(adjNode, visited, stack, adj_list)
        stack.append(current_node)      

    def topoSort(self, V, edges):
        adj_list = [[] for _ in range(V)]
        
        for u, v in edges:
            adj_list[u].append(v)
            
        stack = []
        visited = [0 for _ in range(V)]
        
        for i in range(V):
            if visited[i] == 0:
                self.dfs(i, visited, stack, adj_list)
                
        return stack[::-1]


# Driver code to test
if __name__ == "__main__":
    # Example graph
    V = 6
    edges = [
        [5, 2],
        [5, 0],
        [4, 0],
        [4, 1],
        [2, 3],
        [3, 1]
    ]
    
    obj = Solution()
    result = obj.topoSort(V, edges)
    print("Topological Sort Order:", result)
```
