```python
# clone_graph_dfs.py

# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        # A dictionary to keep track of copied nodes
        visited = {}

        # DFS function to clone each node
        def dfs(curr):
            # If already cloned, return the copy
            if curr in visited:
                return visited[curr]

            # Create a copy of the current node
            copy = Node(curr.val)
            visited[curr] = copy  # Store in visited

            # Recursively clone all neighbors
            for neighbor in curr.neighbors:
                copy.neighbors.append(dfs(neighbor))

            return copy

        # Handle null input
        return dfs(node) if node else None
```

```python
from collections import deque

# Step 1: Node Definition
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

# Step 2: Clone Graph using DFS
class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        visited = {}

        def dfs(curr):
            if curr in visited:
                return visited[curr]

            copy = Node(curr.val)
            visited[curr] = copy

            for neighbor in curr.neighbors:
                copy.neighbors.append(dfs(neighbor))

            return copy

        return dfs(node) if node else None

# Step 3: Build Sample Graph
# Let's build: 1--2
#              |  |
#              4--3

node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.neighbors = [node2, node4]
node2.neighbors = [node1, node3]
node3.neighbors = [node2, node4]
node4.neighbors = [node1, node3]

# Step 4: Clone and Print Graph
def print_graph(start_node):
    visited = set()
    queue = deque([start_node])
    while queue:
        curr = queue.popleft()
        if curr in visited:
            continue
        visited.add(curr)
        print(f"Node {curr.val} has neighbors {[n.val for n in curr.neighbors]}")
        for neighbor in curr.neighbors:
            if neighbor not in visited:
                queue.append(neighbor)

solution = Solution()
cloned = solution.cloneGraph(node1)

print("Original Graph:")
print_graph(node1)

print("\nCloned Graph:")
print_graph(cloned)
```

```python
# clone_graph_bfs.py

from collections import deque

# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        if not node:
            return None

        # Dictionary to store original â†’ clone mapping
        visited = {}
        clone = Node(node.val)
        visited[node] = clone

        # Initialize queue for BFS
        queue = deque([node])

        while queue:
            curr = queue.popleft()

            # Traverse neighbors
            for neighbor in curr.neighbors:
                if neighbor not in visited:
                    # Clone and store the neighbor
                    visited[neighbor] = Node(neighbor.val)
                    queue.append(neighbor)

                # Append the cloned neighbor to the current node's clone
                visited[curr].neighbors.append(visited[neighbor])

        return clone
```
