The "Graph Valid Tree" problem asks you to determine whether a given undirected graph is a valid tree.

Approach: DFS

A graph is a valid tree if:
It is fully connected (you can reach all nodes from any node).
It has no cycles.

Steps:

Build the graph as an adjacency list.
Use DFS to visit all nodes.
If a node is visited again (except the parent), it's a cycle â†’ not a tree.
After DFS, check if all nodes were visited (to confirm it's connected).

Steps:

Edge Count Check:
If number of edges â‰  n - 1, it cannot be a tree â†’ return False.

Build the Graph:
Use an adjacency list to represent the undirected graph.

DFS Traversal:
Start DFS from any node (usually node 0).
Track visited nodes using a visited set.

For each neighbor:
If it's the parent, skip.
If it's already visited â†’ cycle detected â†’ return False.

Final Connectivity Check:
After DFS, ensure the visited set contains all n nodes.
If not â†’ graph is disconnected â†’ return False.

Return True if:
No cycles were found
All nodes are visited

```python
from typing import List

class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False  # A valid tree must have exactly n-1 edges

        # Build adjacency list
        graph = {i: [] for i in range(n)}
        for u, v in edges:
            graph[u].append(v)
            graph[v].append(u)

        visited = set()

        def dfs(node, parent):
            if node in visited:
                return False  # Cycle detected

            visited.add(node)
            for neighbor in graph[node]:
                if neighbor == parent:
                    continue  # Skip the parent node
                if not dfs(neighbor, node):
                    return False
            return True

        # Start DFS from node 0
        if not dfs(0, -1):
            return False

        # All nodes must be connected (i.e., visited)
        return len(visited) == n


# Example usage
sol = Solution()
print(sol.validTree(5, [[0,1],[0,2],[0,3],[1,4]]))  # True
print(sol.validTree(5, [[0,1],[1,2],[2,3],[1,3],[1,4]]))  # False (cycle)
```

Approach: Union-Find (Disjoint Set)

A graph is a tree if and only if:
It is fully connected (all nodes are reachable).
It has no cycles (acyclic).
It has exactly n - 1 edges for n nodes.

ðŸ” Steps:

If number of edges â‰  n - 1 â†’ return False immediately.
Initialize Union-Find (parent array).
For each edge (u, v), union them.
If union finds theyâ€™re already connected â†’ cycle detected â†’ return False.
After all edges processed â†’ return True

from typing import List

class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False

        parent = [i for i in range(n)]

        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])  # Path compression
            return parent[x]

        def union(x, y):
            rootX = find(x)
            rootY = find(y)
            if rootX == rootY:
                return False  # Cycle detected
            parent[rootY] = rootX
            return True

        for u, v in edges:
            if not union(u, v):
                return False

        return True


# Example usage
sol = Solution()
print(sol.validTree(5, [[0,1],[0,2],[0,3],[1,4]]))  # True
print(sol.validTree(5, [[0,1],[1,2],[2,3],[1,3],[1,4]]))  # False
