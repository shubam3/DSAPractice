```python
"""
Definition of Interval:
class Interval(object):
    def __init__(self, start, end):
        self.start = start
        self.end = end
"""

from typing import List

class Solution:
    def canAttendMeetings(self, intervals: List['Interval']) -> bool:
        for i in range(len(intervals)):
            for j in range(i + 1, len(intervals)):
                # Use .start and .end instead of [0], [1]
                if intervals[i].start < intervals[j].end and intervals[j].start < intervals[i].end:
                    return False
        return True
   
```

The idea is to compare every meeting with every other meeting. The outer loop picks one meeting, and the inner loop compares it with all meetings that come after it. For each pair of meetings, we check if their time ranges overlap. The condition intervals[i].start < intervals[j].end and intervals[j].start < intervals[i].end means that the start time of one meeting happens before the end of the other and vice versa, which indicates an overlap. If any overlap is found, it means the person cannot attend both meetings, so the function immediately returns False.

If the loops finish without finding any overlapping intervals, it means all meetings are scheduled at separate times, and the person can attend all of them, so the function returns True. This is a brute-force approach because it compares every pair of meetings, giving it a time complexity of O(n²).

```python
from typing import List

class Solution:
    def canAttendMeetings(self, intervals: List['Interval']) -> bool:
        # Sort intervals by start time
        intervals.sort(key=lambda x: x.start)

        for i in range(1, len(intervals)):
            if intervals[i].start < intervals[i-1].end:
                return False
        return True
```

The optimal approach is to first sort the meeting intervals based on their start time. Once the intervals are sorted, we only need to compare each meeting with the one immediately before it. If the start time of the current meeting is less than the end time of the previous meeting, it means the meetings overlap and the person cannot attend both, so we return false. If we iterate through all intervals without finding any overlap, it means all meetings are scheduled without conflicts and the person can attend them all.

The reason this works is because sorting ensures that if an overlap exists, it will always appear between adjacent intervals. The time complexity is O(n log n) due to sorting, and the space complexity is O(1) aside from the sorting operation.
