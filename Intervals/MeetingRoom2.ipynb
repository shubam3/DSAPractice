🧩 Goal

Given a list of meeting intervals [start, end],
find the minimum number of meeting rooms required so that no two overlapping meetings share the same room.

💡 What You Have to Do

1️⃣ Each meeting is an interval of time.
2️⃣ Whenever meetings overlap in time, they must use different rooms.
3️⃣ Find the maximum number of meetings that overlap at any time — that number = minimum rooms needed.

⚙️ Example
intervals = [[0,30],[5,10],[15,20]]


Timeline view

0-----5-----10----15----20----30
|-----------|                  (0–30)
      |--|                      (5–10)
           |----|               (15–20)


At most 2 meetings overlap → ✅ 2 rooms needed

```python
# meeting_rooms_2_bruteforce.py

from typing import List


class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        n = len(intervals)
        max_rooms = 0

        for i in range(n):
            rooms = 1
            for j in range(n):
                if i != j:
                    # check overlap
                    if intervals[j][0] < intervals[i][1] and intervals[j][1] > intervals[i][0]:
                        rooms += 1
            max_rooms = max(max_rooms, rooms)

        return max_rooms

# -------------------------------
# Example Test Cases
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    # Example 1
    intervals1 = [[0, 30], [5, 10], [15, 20]]
    print("Input:", intervals1)
    print("Minimum Meeting Rooms Needed:", obj.minMeetingRooms(intervals1))
    # Expected Output: 2

    # Example 2
    intervals2 = [[7, 10], [2, 4]]
    print("\nInput:", intervals2)
    print("Minimum Meeting Rooms Needed:", obj.minMeetingRooms(intervals2))
    # Expected Output: 1
```

In the brute force approach for the Meeting Rooms II problem, we compare every meeting with every other meeting to determine how many meetings overlap at a given time. The idea is that if meetings overlap, they must be held in different rooms.

For each meeting, we treat it as a reference and check all other meetings to see if their time intervals intersect with it. Two meetings overlap if the start time of one meeting is less than the end time of the other and the end time of the first meeting is greater than the start time of the other. Every time we detect an overlap, we increase a counter that represents the number of rooms required at that moment.

After checking all meetings against the current one, we update the maximum number of overlapping meetings we have seen so far. This maximum value represents the minimum number of rooms needed to hold all meetings without conflicts.

However, this approach is inefficient because it compares every pair of meetings, resulting in a time complexity of O(n²). While it is easy to implement and helps understand the overlap concept, more optimal solutions use sorting with a min-heap or sweep-line technique to reduce the complexity to O(n log n).

Optimal Approach (Min-Heap / Priority Queue)
Idea

We want the minimum number of rooms so that no two overlapping meetings share a room.

Key observation:

At any time, the number of rooms needed = number of meetings happening at the same time.

We can track this using a min-heap of end times:

Steps (for notes)

1️⃣ Sort the intervals by start time

So we process meetings in the order they begin.

2️⃣ Use a min-heap to store end times of ongoing meetings

heap[0] (top) = the earliest meeting to finish.

3️⃣ For each meeting [start, end]:

If start >= heap[0]:
→ The earliest meeting has already ended → reuse that room → heap.pop().

Push the current meeting’s end into the heap → it occupies a room.

4️⃣ After processing all meetings:

The size of the heap = number of rooms needed (max parallel meetings).

Complexity

Time: O(n log n) (sorting + heap operations)

Space: O(n) for the heap

```python
# meeting_rooms_2_optimal.py

from typing import List
import heapq


class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0

        intervals.sort(key=lambda x: x[0])  # sort by start time
        minHeap = []  # stores end times

        for start, end in intervals:
            if minHeap and start >= minHeap[0]:
                heapq.heappop(minHeap)  # reuse room

            heapq.heappush(minHeap, end)  # allocate room / update end time

        return len(minHeap)


# -------------------------------
# Example usage
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    intervals1 = [[0, 30], [5, 10], [15, 20]]
    print("Input:", intervals1)
    print("Min rooms needed:", obj.minMeetingRooms(intervals1))  # Expected: 2

    intervals2 = [[7, 10], [2, 4]]
    print("\nInput:", intervals2)
    print("Min rooms needed:", obj.minMeetingRooms(intervals2))  # Expected: 1
```

The optimal approach for the Meeting Rooms II problem uses sorting and a min-heap. First, we sort all meeting intervals based on their start times. This ensures that we process meetings in chronological order. Then we use a min-heap to keep track of the end times of meetings that are currently using rooms.

For each meeting, we check the earliest ending meeting stored at the top of the heap. If the current meeting’s start time is greater than or equal to that end time, it means that room has become free, so we remove it from the heap and reuse the room. Otherwise, the meeting overlaps with existing ones, and we need to allocate a new room.

After that, we push the current meeting’s end time into the heap to indicate that the room is occupied until that time. The size of the heap at any point represents the number of rooms currently being used. After processing all meetings, the heap size gives us the minimum number of meeting rooms required.

The time complexity is O(n log n) due to sorting and heap operations, and the space complexity is O(n) in the worst case when all meetings overlap.
