ðŸ§© Goal

Given a list of meeting intervals [start, end],
find the minimum number of meeting rooms required so that no two overlapping meetings share the same room.

ðŸ’¡ What You Have to Do

1ï¸âƒ£ Each meeting is an interval of time.
2ï¸âƒ£ Whenever meetings overlap in time, they must use different rooms.
3ï¸âƒ£ Find the maximum number of meetings that overlap at any time â€” that number = minimum rooms needed.

âš™ï¸ Example
intervals = [[0,30],[5,10],[15,20]]


Timeline view

0-----5-----10----15----20----30
|-----------|                  (0â€“30)
      |--|                      (5â€“10)
           |----|               (15â€“20)


At most 2 meetings overlap â†’ âœ… 2 rooms needed

```python
# meeting_rooms_2_bruteforce.py

from typing import List

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        # If there are no meetings, no room is needed
        if not intervals:
            return 0

        n = len(intervals)
        max_rooms = 0  # This will store the maximum overlap (rooms needed)

        # Helper function to check if two meetings overlap
        def overlap(a, b):
            # Two intervals overlap if one's start is before the other's end and vice versa
            return not (a[1] <= b[0] or b[1] <= a[0])

        # For each meeting, count how many other meetings overlap with it
        for i in range(n):
            current_overlap = 1  # Count the current meeting itself
            for j in range(n):
                if i == j:
                    continue  # Skip comparing the same meeting
                if overlap(intervals[i], intervals[j]):
                    current_overlap += 1  # Found an overlap, need another room
            # Track the maximum overlap found so far
            max_rooms = max(max_rooms, current_overlap)

        return max_rooms


# -------------------------------
# Example Test Cases
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    # Example 1
    intervals1 = [[0, 30], [5, 10], [15, 20]]
    print("Input:", intervals1)
    print("Minimum Meeting Rooms Needed:", obj.minMeetingRooms(intervals1))
    # Expected Output: 2

    # Example 2
    intervals2 = [[7, 10], [2, 4]]
    print("\nInput:", intervals2)
    print("Minimum Meeting Rooms Needed:", obj.minMeetingRooms(intervals2))
    # Expected Output: 1
```

```python
# meeting_rooms_2_twoptr.py

from typing import List

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0

        starts = sorted(s for s, _ in intervals)
        ends   = sorted(e for _, e in intervals)

        i = j = 0
        rooms = max_rooms = 0

        while i < len(starts):
            if starts[i] < ends[j]:
                rooms += 1
                max_rooms = max(max_rooms, rooms)
                i += 1
            else:
                rooms -= 1
                j += 1

        return max_rooms


if __name__ == "__main__":
    obj = Solution()
    print(obj.minMeetingRooms([[0,30],[5,10],[15,20]]))  # 2
    print(obj.minMeetingRooms([[7,10],[2,4]]))           # 1
```

Optimal Approach (Min-Heap / Priority Queue)
Idea

We want the minimum number of rooms so that no two overlapping meetings share a room.

Key observation:

At any time, the number of rooms needed = number of meetings happening at the same time.

We can track this using a min-heap of end times:

Steps (for notes)

1ï¸âƒ£ Sort the intervals by start time

So we process meetings in the order they begin.

2ï¸âƒ£ Use a min-heap to store end times of ongoing meetings

heap[0] (top) = the earliest meeting to finish.

3ï¸âƒ£ For each meeting [start, end]:

If start >= heap[0]:
â†’ The earliest meeting has already ended â†’ reuse that room â†’ heap.pop().

Push the current meetingâ€™s end into the heap â†’ it occupies a room.

4ï¸âƒ£ After processing all meetings:

The size of the heap = number of rooms needed (max parallel meetings).

Complexity

Time: O(n log n) (sorting + heap operations)

Space: O(n) for the heap

```python
# meeting_rooms_2_optimal.py

from typing import List
import heapq

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        # If there are no meetings, no room is needed
        if not intervals:
            return 0

        # Step 1: Sort meetings by start time
        intervals.sort(key=lambda x: x[0])

        # Min-heap to store end times of ongoing meetings
        min_heap = []  # each element = meeting end time

        # Step 2: Process each meeting
        for start, end in intervals:
            # If the earliest ending meeting ends before or at current start,
            # we can reuse that room (remove it from heap)
            if min_heap and min_heap[0] <= start:
                heapq.heappop(min_heap)

            # Allocate current meeting's end time into a room (heap)
            heapq.heappush(min_heap, end)

        # Step 3: Heap size = number of rooms needed
        return len(min_heap)


# -------------------------------
# Example usage
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    intervals1 = [[0, 30], [5, 10], [15, 20]]
    print("Input:", intervals1)
    print("Min rooms needed:", obj.minMeetingRooms(intervals1))  # Expected: 2

    intervals2 = [[7, 10], [2, 4]]
    print("\nInput:", intervals2)
    print("Min rooms needed:", obj.minMeetingRooms(intervals2))  # Expected: 1
```
