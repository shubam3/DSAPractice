Initialize:

Create an empty list res to store the result.

2️⃣ Iterate through each interval:

Case 1: If newInterval ends before the current interval starts
→ append newInterval and return res + remaining intervals.

Case 2: If newInterval starts after the current interval ends
→ append the current interval to res.

Case 3: If they overlap, merge them:

newInterval[0] = min(newInterval[0], intervals[i][0])
newInterval[1] = max(newInterval[1], intervals[i][1])


3️⃣ After the loop:

Append the last newInterval to res.

4️⃣ Return res.

The intervals are already sorted and non-overlapping, so we can process them in one pass. The idea is to iterate through each interval and compare it with the new interval we want to insert. There are three possible cases.

First, if the new interval ends before the current interval starts, it means the new interval should appear before the current one and there will be no overlap with any remaining intervals. So we append the new interval to the result and return the rest of the intervals.

Second, if the current interval ends before the new interval starts, there is no overlap and the current interval should appear before the new interval. In that case, we simply add the current interval to the result and continue.

The third case is when the intervals overlap. Instead of adding anything immediately, we merge the intervals by updating the new interval’s start to the minimum of the two starts and the end to the maximum of the two ends. This effectively expands the new interval to absorb the overlapping interval.

After processing all intervals, we append the final merged new interval to the result. Since we traverse the list only once, the time complexity is O(n) and the space complexity is O(n) for the result list.

```python
from typing import List

class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        res = []  # result list to store merged intervals

        # Loop through all existing intervals
        for i in range(len(intervals)):

            # Case 1️⃣: newInterval lies completely before the current interval
            # Example: newInterval = [1, 2], current interval = [3, 5]
            if newInterval[1] < intervals[i][0]:
                # Append newInterval, then append the rest and return directly
                res.append(newInterval)
                return res + intervals[i:]

            # Case 2️⃣: newInterval lies completely after the current interval
            # Example: current interval = [1, 2], newInterval = [3, 5]
            elif newInterval[0] > intervals[i][1]:
                res.append(intervals[i])  # add the current non-overlapping interval

            # Case 3️⃣: Overlapping intervals
            # Merge newInterval with current interval
            else:
                newInterval = [
                    min(newInterval[0], intervals[i][0]),  # start = smaller of the two
                    max(newInterval[1], intervals[i][1])   # end = larger of the two
                ]

        # After loop, append the last newInterval
        res.append(newInterval)
        return res


# -------------------------------
# Example Test
# -------------------------------
if __name__ == "__main__":
    obj = Solution()
    intervals = [[1, 3], [6, 9]]
    newInterval = [2, 5]
    print("Output:", obj.insert(intervals, newInterval))
    # Expected Output: [[1, 5], [6, 9]]
```
