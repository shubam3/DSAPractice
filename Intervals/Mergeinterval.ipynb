```python
# merge_intervals.py

from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # Step 1: Sort intervals by their start time
        intervals.sort(key=lambda i: i[0])
        output = [intervals[0]]  # Initialize output with the first interval

        # Step 2: Iterate through remaining intervals
        for start, end in intervals[1:]:
            lastEnd = output[-1][1]  # End of last merged interval

            # Case 1️⃣: Overlapping intervals
            if start <= lastEnd:
                # Merge by updating the end of the last interval
                output[-1][1] = max(lastEnd, end)
            else:
                # Case 2️⃣: No overlap → add interval as new
                output.append([start, end])

        return output


# -------------------------------
# Example Test
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    intervals = [[1, 3], [2, 6], [8, 10], [15, 18]]
    result = obj.merge(intervals)
    print("Merged Intervals:", result)
    # Expected Output: [[1, 6], [8, 10], [15, 18]]
```
