Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

 

Example 1:

Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
Explanation: Since intervals [1,3] and [2,6] overlap, merge them into [1,6].

```python
# merge_intervals.py

from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # Step 1: Sort intervals by their start time
        intervals.sort(key=lambda i: i[0])
        output = [intervals[0]]  # Initialize output with the first interval

        # Step 2: Iterate through remaining intervals
        for start, end in intervals[1:]:
            lastEnd = output[-1][1]  # End of last merged interval

            # Case 1️⃣: Overlapping intervals
            if start <= lastEnd:
                # Merge by updating the end of the last interval
                output[-1][1] = max(lastEnd, end)
            else:
                # Case 2️⃣: No overlap → add interval as new
                output.append([start, end])

        return output


# -------------------------------
# Example Test
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    intervals = [[1, 3], [2, 6], [8, 10], [15, 18]]
    result = obj.merge(intervals)
    print("Merged Intervals:", result)
    # Expected Output: [[1, 6], [8, 10], [15, 18]]
```

This solution solves the Merge Intervals problem using an optimal sorting and single-pass approach. First, I sort the intervals based on their starting times so that any overlapping intervals appear next to each other. Then, I initialize the output list with the first interval. As I iterate through the remaining intervals, I compare the current interval’s start time with the end time of the last interval in the output list. If the current start is less than or equal to the last end, it means the intervals overlap, so I merge them by updating the end to the maximum of both end values. If they do not overlap, I simply append the current interval to the output list. This ensures each interval is processed once after sorting. The time complexity is O(n log n) due to sorting, and the space complexity is O(n) for storing the merged intervals.
