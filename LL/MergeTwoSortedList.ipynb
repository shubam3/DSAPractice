```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
            dummy = ListNode()
            tail = dummy
            l1, l2 = list1, list2

            while l1 and l2:
                if l1.val < l2.val:
                    tail.next = l1
                    l1 = l1.next
                else:
                    tail.next = l2
                    l2 = l2.next
                tail = tail.next

            tail.next = l1 if l1 else l2
            return dummy.next


# 1️⃣ Create dummy node: dummy → start of new merged list

# tail → pointer to the last node of merged list

# 2️⃣ Point to both lists
# 3️⃣ Traverse both lists
# 4️⃣ Compare values
# Case 1: l1 smaller : Attach node from list1.
# Case 2: l2 smaller : Attach node from list2.
# Move tail forward
# 5️⃣ Add remaining nodes : If one list ends, attach the remaining nodes.
# 6️⃣ Return merged list
```

We create a dummy node to build the merged list. Then we compare the current nodes of both lists and attach the smaller one to the result. We move the pointer of that list forward and also move the tail pointer. Once one list ends, we attach the remaining nodes of the other list. Finally, we return dummy.next.
