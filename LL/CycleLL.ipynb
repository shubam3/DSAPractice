```python
def hasCycle(self, head: Optional[ListNode]) -> bool:
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next          # move 1 step
        fast = fast.next.next     # move 2 steps

        if slow == fast:
            return True           # cycle detected

    return False                  # no cycle


# slow moves one step at a time.

# fast moves two steps at a time.

# If there is a cycle, fast will eventually meet slow.

# If fast reaches None, there is no cycle.
```

To detect a cycle in a linked list, I use Floyd’s Cycle Detection algorithm, also known as the Tortoise and Hare approach. I maintain two pointers: a slow pointer that moves one step at a time and a fast pointer that moves two steps at a time. If there is no cycle, the fast pointer will eventually reach the end of the list. However, if a cycle exists, the fast pointer will eventually catch up to the slow pointer inside the loop because it is moving faster. When both pointers meet, we confirm that a cycle is present. This approach runs in O(n) time and uses O(1) space since no extra data structures are needed.
