```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        values = []

        # Step 1: Collect all node values
        for node in lists:
            while node:
                values.append(node.val)
                node = node.next

        # Step 2: Sort the values
        values.sort()

        # Step 3: Build new sorted linked list
        dummy = ListNode()
        current = dummy
        for val in values:
            current.next = ListNode(val)
            current = current.next

        return dummy.next
```

his solution merges k sorted linked lists using a simple brute-force approach. First, we traverse every linked list and collect all node values into a single array. This step takes O(N) time, where N is the total number of nodes across all lists. Next, we sort the collected values, which takes O(N log N) time. Finally, we build a new linked list by iterating through the sorted array and creating new nodes in order. This guarantees a correctly merged sorted list. While this approach is easy to implement and understand, it is not the most efficient because sorting all elements takes more time compared to the heap-based solution, which runs in O(N log k). The space complexity here is O(N) since we store all node values in an array.
