To solve the Product of Array Except Self problem without using division, I use a prefix and suffix product approach. The idea is that for each index, the result should be the product of all elements to the left multiplied by the product of all elements to the right. In the first pass, I compute prefix products and store them in the result array, where result[i] contains the product of all elements before index i. In the second pass, I traverse from right to left while maintaining a suffix product and multiply it with the existing prefix value in the result array. This way, each position gets left × right product without including itself. The time complexity is O(n) and the space complexity is O(1) extra space since we reuse the output array.

```python
def productExceptSelf(nums):
            n = len(nums)
            res = [1] * n

            # Left products
            prefix = 1
            for i in range(n):
                res[i] = prefix
                prefix *= nums[i]

            # Right products
            suffix = 1
            for i in range(n - 1, -1, -1):
                res[i] *= suffix
                suffix *= nums[i]

            return res
```

```python
# Input: nums = [1,2,4,6]
# Output: [48,24,12,8]

class Solution:
    def productitself(self, nums: List[int]) -> List[int]:
        n= len(nums)
        output = [1] * n

        for i in range(n):
            product = 1 #intialize product for the current index 
            for j in range(n):
                if i!=j:
                    product *= nums[j] 
                output[i] = product
        return output            
```
