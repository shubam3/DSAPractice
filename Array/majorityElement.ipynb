Find the majority element → the number that appears more than n/2 times.

nums = [2, 2, 1, 1, 1, 2, 2]
n = 7
n // 2 = 3
So any number appearing more than 3 times is the answer.

```python
def majorityElement(nums):
    n = len(nums)

    for i in range(n):
        count = 0
        for j in range(n):
            if nums[j] == nums[i]:
                count += 1

        if count > n // 2:
            return nums[i]
```

```python
# We create a dictionary (frequency) to count how many times each number appears.

# We loop through the array one number at a time.

# For each number:

# Increase its count in the dictionary.

# If its count becomes the highest so far, update the majority element.

# At the end, return the number that appeared the most.

       
def majorityElement(nums):
    frequency = {}
    max_count = 0
    majority = None

    for num in nums:
        frequency[num] = frequency.get(num, 0) + 1

        if frequency[num] > max_count:
            max_count = frequency[num]
            majority = num

    return majority


# ---- Test the function ----
nums = [2, 2, 1, 1, 1, 2, 2]
print("Majority Element:", majorityElement(nums))
```

```python
# Pair one majority element with one non-majority element
# Cancel them out
# Majority element will still remain at the end

from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        candidate = None
        vote_count = 0

        for number in nums:
            # If no current candidate, choose this number
            if vote_count == 0:
                candidate = number

            # Vote for or against the candidate
            if number == candidate:
                vote_count += 1
            else:
                vote_count -= 1

        return candidate


# --------- Run & Test in VS Code ----------
if __name__ == "__main__":
    nums = [2, 2, 1, 1, 1, 2, 2]
    solution = Solution()
    result = solution.majorityElement(nums)
    print("Majority Element:", result)


                   
```
