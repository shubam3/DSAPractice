You are given:

An integer array nums
An integer k

You must return the k most frequent elements in the array.

Example:

Input:
nums = [1,1,1,2,2,3], k = 2
Output:
[1,2]
→ Because 1 appears 3 times and 2 appears 2 times.

```python
# Counts frequency of each number
# Sorts based on frequency (descending)
# Returns the top k frequent numbers

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # Step 1: Extract unique elements
        unique_nums = []
        frequencies = []

        for num in nums:
            if num in unique_nums:
                index = unique_nums.index(num)
                frequencies[index] += 1  # Increment frequency
            else:
                unique_nums.append(num)
                frequencies.append(1)  # Initialize frequency

        # Step 2: Sort unique_nums based on frequencies
        for i in range(len(frequencies)):
            for j in range(i + 1, len(frequencies)):
                if frequencies[i] < frequencies[j]:  # Sort in descending order
                    frequencies[i], frequencies[j] = frequencies[j], frequencies[i]
                    unique_nums[i], unique_nums[j] = unique_nums[j], unique_nums[i]

        # Step 3: Return the top k elements
        return unique_nums[:k]  # Take the first k elements
```

Time Complexity

1. Building unique_nums and frequencies

For every element in nums, you do:
num in unique_nums → O(n) search
unique_nums.index(num) → another O(n) search

This happens n times.
So this part is:
O(n²)

2. Nested loop sorting

Two nested loops:

for i in range(len(frequencies)):
    for j in range(i + 1, len(frequencies)):


Worst-case: unique numbers = n
So sorting takes:
O(n²)

3. Slicing unique_nums[:k]

This takes:

O(k) (very small compared to n²)
Final Time Complexity
🎯 O(n²) + O(n²) + O(k) → O(n²)

Overall Time Complexity: O(n²)

This is why it’s considered a brute-force solution.

✅ Space Complexity

You store:
unique_nums → at most n unique elements
frequencies → at most n integers

So space is:
O(n)

No extra data structures apart from these.

🎯 Final Answer
Part	Complexity
Time Complexity	O(n²)
Space Complexity	O(n)

```python
# Yup, let’s do the optimal one (bucket sort, O(n) average).

# Idea (simple)

# Count frequency of each number.
# Make “buckets” where index = frequency and value = list of numbers.
# Traverse buckets from high freq to low and collect k numbers.


from typing import List
from collections import Counter

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # 1. Count frequencies
        freq = Counter(nums)              # num -> count

        # 2. Buckets: index = frequency, value = list of nums
        buckets = [[] for _ in range(len(nums) + 1)]
        for num, count in freq.items():
            buckets[count].append(num)

        # 3. Traverse buckets from right (max freq) to left
        result = []
        for f in range(len(buckets) - 1, 0, -1):
            for num in buckets[f]:
                result.append(num)
                if len(result) == k:
                    return result
```

1. Counting Frequency → O(n)
freq = Counter(nums)


Counter goes through each element once → O(n).

✅ 2. Creating Buckets → O(n)
buckets = [[] for _ in range(len(nums) + 1)]

We create n + 1 empty lists → O(n).
Then we insert each unique number into its bucket:

for num, count in freq.items():
    buckets[count].append(num)


There are at most n unique numbers
Each number is inserted once → O(n)

✅ 3. Scanning Buckets from end → O(n)
for f in range(len(buckets)-1, 0, -1):
    for num in buckets[f]:
        ...

Why this is O(n)?

Total numbers across all buckets = n
We never touch more than n elements

We stop early when we collect k elements
So in the worst case, we inspect each bucket element once → O(n).

🎯 Total Time = O(n) + O(n) + O(n) → O(n)

We only do a few loops that depend linearly on the length of the array → O(n).

```python
from typing import List
from collections import Counter
import heapq

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # Step 1: Count frequencies
        freq = Counter(nums)  # hashmap: num -> frequency
        
        # Step 2: Use a min-heap of size k
        min_heap = []  # stores (frequency, number)

        for num, count in freq.items():
            heapq.heappush(min_heap, (count, num))  # push pair

            # If heap grows larger than k, remove smallest frequency
            if len(min_heap) > k:
                heapq.heappop(min_heap)

        # Step 3: Extract results (only top k remain in heap)
        result = []
        for count, num in min_heap:
            result.append(num)

        return result
```

✅ 1. Bucket Sort Approach (Optimal O(n))
Approach

Count frequencies of all elements using a hashmap (num → frequency).

Create a bucket list where each index represents a frequency.

Example: buckets[3] contains all numbers that appear 3 times.

Place each number into the bucket corresponding to its frequency.

Traverse buckets from highest frequency to lowest and collect elements until you have k results.

Return the collected top k numbers.

Why it is O(n)

Counting frequencies → O(n)

Building buckets → O(n)

Scanning buckets once → O(n)
→ Total O(n)

✅ 2. Min-Heap Approach (O(n log k))
Approach

Count frequencies of all elements using a hashmap.

Use a min-heap of size k to store the most frequent elements.

For each (num, freq) pair:

Push into heap

If heap size exceeds k, remove the element with the lowest frequency

After processing all numbers, the heap contains the k most frequent elements.

Extract the numbers from the heap and return them.

Why it is O(n log k)

Inserting each unique number into heap → O(log k)

Done for up to n numbers → O(n log k)
