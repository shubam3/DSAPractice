```python
# Output: 6
# Explanation: The subarray [4, -1, 2, 1] has the maximum sum 6

def max_subarray_sum(arr):
  
    # Implement the function
    if not arr:
        return 0
 
    # # Initialize variables
    max_ending_here = 0
    max_so_far = float('-inf')
 
    # # Iterate through the array
    for x in arr:
        # Update max_ending_here to include current element
        max_ending_here = max(x, max_ending_here + x)
        # Update max_so_far to the maximum value found so far
        max_so_far = max(max_so_far, max_ending_here)
 
    return max_so_far
```

To find the maximum sum of a contiguous subarray efficiently, I use Kadane’s Algorithm, which is a dynamic programming approach. The key idea is that if the current running sum becomes negative, it cannot contribute to a larger future sum, so we reset it to zero. As we iterate through the array, we keep adding elements to a running sum and continuously update the maximum sum seen so far. At each step, we decide whether to extend the current subarray or start fresh. This ensures we process each element only once, giving us an O(n) time complexity and O(1) space complexity.

```python
#Kadaen's Algorithm: Kadane’s Algorithm is an efficient dynamic programming approach to 
    #find the maximum sum of a contiguous subarray in an array.
def maxsubarray(arr):

    if not arr:  # Check if the array is empty
        return 0  
    
    maxSum = arr[0]
    currSum = 0
  
    
    for n in arr:
        if currSum <0:
           currSum = 0
          
        currSum += n
        maxSum = max(maxSum,currSum)
         
    return maxSum  
```
