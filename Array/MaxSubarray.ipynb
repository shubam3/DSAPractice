This approach solves the Maximum Subarray problem using a brute-force technique. I use two nested loops where the first loop selects a starting index, and the second loop expands the subarray from that starting point while continuously adding elements to a running sum. At each step, I compare the current subarray sum with the maximum sum found so far and update it if needed. This way, we check all possible subarrays to ensure we don’t miss the maximum one. The time complexity is O(n²) because of the nested loops, and the space complexity is O(1) since only variables are used.

```python
# Output: 6
# Explanation: The subarray [4, -1, 2, 1] has the maximum sum 6

def maxSubArray(nums):
    n = len(nums)
    maxSum = float('-inf')

    for i in range(n):
        currSum = 0
        for j in range(i, n):
            currSum += nums[j]
            maxSum = max(maxSum, currSum)

    return maxSum
```
