Approach (Sorting-Based, O(n log n))

If the list is empty → return 0.
Sort the array so consecutive numbers appear next to each other.
Maintain two variables:
current_streak: current consecutive sequence length
longest_streak: longest sequence seen so far

Loop through the sorted list:

If the current number equals the previous number + 1 → extend streak.
If the current number equals the previous number → duplicate → ignore.

Otherwise, reset the streak to 1.
After the loop, return the larger of longest_streak or current_streak

```python
def longestConsecutive(nums):
   
    if not nums:   # O(n logn ) because of string
        return 0

    nums.sort()
    longest_streak = 1
    current_streak = 1

    for i in range(1, len(nums)):
        if nums[i] != nums[i-1]:
            if nums[i] == nums[i-1] + 1:
                current_streak += 1
            else:
                longest_streak = max(longest_streak, current_streak)
                current_streak = 1

    return max(longest_streak, current_streak)

```

Optimal O(n) Solution — Hash Set

Insert numbers into a set.

A number is a “start” if num - 1 is not in the set.

From each start, count consecutive numbers using a while loop.

Track the maximum streak.

Only O(n) total work because each number is visited at most once

```python
def longestConsecutive(nums):
    num_set = set(nums)
    longest_streak = 0

    for num in num_set:
        # check if it's the start of a sequence
        if num - 1 not in num_set:
            current_num = num
            current_streak = 1

            # expand sequence
            while current_num + 1 in num_set:
                current_num += 1
                current_streak += 1

            longest_streak = max(longest_streak, current_streak)

    return longest_streak
```

We only count sequences starting points, so the inner while loop runs at most n times total, not n times per element → still O(n).
