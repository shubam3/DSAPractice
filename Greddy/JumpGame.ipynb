You’re given an array nums where nums[i] represents the maximum jump length from that index.
Return True if you can reach the last index; otherwise False.

⚙️ Algorithm

1️⃣ Initialize

max_index = 0
→ the farthest index you can currently reach.

2️⃣ Iterate through the array
For each index i in 0 … len(nums)-1:

Step 1:

If i > max_index, it means you’ve reached a position that you cannot jump to, so return False.

Step 2:

Update the farthest reachable position:
max_index = max(max_index, i + nums[i])


3️⃣ End condition

If you finish the loop without returning False,
→ it means you can always reach or pass the last index → return True.

🧮 Example

nums = [2, 3, 1, 1, 4]

i	nums[i]	max_index before	i > max_index?	new max_index = max(old, i+nums[i])
0	2	0	No	2
1	3	2	No	4
2	1	4	No	4
3	1	4	No	4
4	4	4	No	8

✅ Reached end → return True

❌ Example 2

nums = [3,2,1,0,4]

At i=3, max_index=3, and i+nums[i]=3, can’t go further.
When i=4, i > max_index → return False.

⏱️ Complexity

Time: O(n)

Space: O(1)

```python
# jump_game.py

from typing import List

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        # max_index represents the farthest position we can currently reach
        max_index = 0

        # Traverse each index in the array
        for i in range(len(nums)):
            # If the current index is beyond the farthest we can reach,
            # it means we got stuck and cannot move further
            if i > max_index:
                return False

            # Update the farthest index reachable so far
            # i + nums[i] means: from index i, you can jump up to i + nums[i]
            max_index = max(max_index, i + nums[i])

        # If we finish the loop, it means we could reach (or pass) the last index
        return True


# -------------------------------
# Example test
# -------------------------------
if __name__ == "__main__":
    obj = Solution()

    nums1 = [2, 3, 1, 1, 4]   # reachable -> True
    nums2 = [3, 2, 1, 0, 4]   # not reachable -> False

    print("Example 1:", obj.canJump(nums1))  # Output: True
    print("Example 2:", obj.canJump(nums2))  # Output: False
```
